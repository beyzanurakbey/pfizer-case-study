```python
import gurobipy as gp
from gurobipy import GRB
import openpyxl# Load the Excel workbook
import matplotlib.pyplot as plt
import numpy as np

workbook = openpyxl.load_workbook('Pfizer Case Dataset.xlsx')
# Create a new model
model = gp.Model("Pfizer Model")
model.ModelSense = GRB.MAXIMIZE


# Naming the worksheets for better reference
worksheet_name1 = 'workload index'
worksheet_name2 = 'current structure'
worksheet_name3 = 'distance'




workloadindexes = workbook[worksheet_name1]
currentstructure = workbook[worksheet_name2]
distances = workbook[worksheet_name3]

wli = {}
for i in range(2, workloadindexes.max_row + 1):
    #print(workloadindexes.cell(row=i, column= 2).value)
    wli.update({i - 1 : workloadindexes.cell(row=i, column= 2).value})
print(wli)

currentallocation = {}

for j in range(2 , currentstructure.max_column + 1 ):
    for k in range(3, currentstructure.max_row + 1):
        
        currentallocation.update({"S{}_{}".format(j-1, k-2): currentstructure.cell(row=k, column= j).value})
    
print(currentallocation)

distance = {}

for j in range(2 , distances.max_column + 1 ):
    for k in range(3, distances.max_row + 1):
        
        distance.update({"S{}_{}".format(j-1, k-2): distances.cell(row=k, column= j).value})
print(distance)  

def pfizermodel(wli, currentallocation, distance, epsilon, alpha):
    print(wli)
    # Create a new model
    model = gp.Model("pfizer model")
    model.ModelSense = GRB.MINIMIZE
    xsb = []
    
    for j in range(0 , 4 ):
        for k in range(0, 22):
        
             x = model.addVar(vtype=gp.GRB.BINARY, name=f"X{j + 1}_{k + 1}")
             xsb.append(x)
    xsb[0] == 1
    xsb[35] == 1
    xsb[59] == 1
    xsb[87] == 1         
    obj_expression = 0
    for m in range(0 , 4 ):
        for l in range(0, 22):
            obj_expression += xsb[m * 22 + l] * distance["S{}_{}".format(m+1, l+1)]        
    csd = []
    for t in range(0, 4): 
        for y in range(0 , 22):
            csd.append(currentallocation["S{}_{}".format(m+1, l+1)])
    
    const1 = 0
    const1 = sum(sum(wli[e+1] * xsb[i * 22 + e] *(1 - currentallocation[f"S{1 + i}_{e + 1}"] )
            for e in range(0, 22))for i in range(0 , 4)) <=  epsilon

    model.addConstr(const1, name='your_constraint_name')                      
  

                                                                 
    for z in range(0, 22):
        xsbconst = 0
        for y in range(0, 4):
            xsbconst += xsb[y * 22 + z]  # Correct indexing for xsb
        model.addConstr(xsbconst == 1)

                                                                  
    for i in range(0, 4):
        wlconst = 0
        for j in range(0, 22):
            wlconst += xsb[i * 22 + j] * wli[j + 1]  # Correct indexing for wlconst
        model.addConstr(wlconst >= 1 - alpha)
        model.addConstr(wlconst <=  1 + alpha)
    
                                                                 
    model.setObjective(obj_expression, GRB.MINIMIZE)
    model.optimize()
    
    #for var in xsb:
     #   print(f"{var.varName}: {var.x}")
    
    
    return model.ObjVal

results = []
for epsilon in np.arange(0.3, 4, 0.1):  # Incrementing by 0.1
    epsilon = epsilon # Adjust epsilon calculation here if needed
    result = pfizermodel(wli, currentallocation, distance, epsilon, 0.1)
    results.append(result)

# Plotting the results
plt.plot(np.arange(0.3, 4, 0.1), results, marker='o')
plt.xlabel('Epsilon')
plt.ylabel('Objective Value')
plt.title('Objective Value vs Epsilon')
plt.grid(True)
plt.show()
    
```
